```python
import numpy as np

plt.hist(df['last_review'])

Q1 = np.percentile(df['last_review'], 25)
Q3 = np.percentile(df['last_review'], 75)
IQR = Q3 - Q1

threshold = 1.5 * IQR

outliers = df['last_review'][(df['last_review'] < Q1 - threshold) | (df['last_review'] > Q3 + threshold)]

plt.scatter(outliers, np.zeros_like(outliers), color='green', label='Outliers')

plt.legend()
plt.show()
```
